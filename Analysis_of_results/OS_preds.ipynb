```python
for therapy in ['PDS', 'IDS']:
    
    if therapy == 'IDS':
        chosen_cancer1_results = ['Cancer_HRP_CCNE1amp_IDS'] 
        number_of_features = 7
    elif therapy == 'PDS':
        chosen_cancer1_results = ['Cancer_HRP_CCNE1amp_PDS']
        number_of_features = 6
    
    result_df = pd.DataFrame(columns=['CellType', 'Marker'])

    # Process Cancer1 Results
    result_df = analysis_of_OS_results.process_results(chosen_cancer1_results, config['cancer1_folder_path'], 'Cancer', result_df, number_of_features, 'heatmap')
    analysis_of_OS_results.create_heatmap_mean_value_per_marker(cells_df, therapy, result_df)
```

### Oncogrid with top five predictive features and their MDA scores across four cell types: All.cells, Cancer.full.panel, Cancer, Immune-stromal.

Plot displays the top five predictive features for comparing short-term and long-term survival within HRDs and HRPs, categorised by cell type. This plot is generated for each therapy sequence. Each grid square indicates the feature's importance based on the median MDA score across 30 splits for a specific comparison within a particular cell type. Squares shaded in grey indicate features removed from the corresponding analysis.

```python
for therapy in ['PDS', 'IDS']:
    
    if therapy == 'IDS':
        chosen_cancer1_results = ['Cancer_BRCAmutmet_HRD_IDS','Cancer_HRP_CCNE1amp_IDS']
        chosen_cancer2_results = ['Cancer_BRCAmutmet_HRD_IDS','Cancer_HRP_CCNE1amp_IDS']
        chosen_non_cancer_results = ['Immune-stromal_BRCAmutmet_HRD_IDS','Immune-stromal_HRP_CCNE1amp_IDS']
        chosen_all_results = ['All_BRCAmutmet_HRD_IDS', 'All_HRP_CCNE1amp_IDS']
    elif therapy == 'PDS':
        chosen_cancer1_results = ['Cancer_BRCAmutmet_HRD_PDS','Cancer_HRP_CCNE1amp_PDS']
        chosen_cancer2_results = ['Cancer_BRCAmutmet_HRD_PDS','Cancer_HRP_CCNE1amp_PDS']
        chosen_non_cancer_results = ['Immune-stromal_BRCAmutmet_HRD_PDS','Immune-stromal_HRP_CCNE1amp_PDS']
        chosen_all_results = ['All_BRCAmutmet_HRD_PDS', 'All_HRP_CCNE1amp_PDS']
    
    result_df = pd.DataFrame()
    
    # Process All Results
    result_df = analysis_of_OS_results.process_results(chosen_all_results, config['all_folder_path'], 'All', result_df, 5, 'oncogrid')

    # Process Cancer1 Results
    result_df = analysis_of_OS_results.process_results(chosen_cancer1_results, config['cancer1_folder_path'], 'Cancer1', result_df, 5, 'oncogrid')

    # Process Cancer2 Results
    result_df = analysis_of_OS_results.process_results(chosen_cancer2_results, config['cancer2_folder_path'], 'Cancer2', result_df, 5, 'oncogrid')

    # Process Non-Cancer Results
    result_df = analysis_of_OS_results.process_results(chosen_non_cancer_results, config['non_cancer_folder_path'], 'Non-cancer', result_df, 5, 'oncogrid')
    
    analysis_of_OS_results.plot_oncogrid_of_pred_features(result_df, therapy)
```
